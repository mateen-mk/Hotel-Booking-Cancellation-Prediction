```python
resort = df[(df['hotel'] == 'Resort Hotel') & (df['is_canceled'] == '0')]
city = df[(df['hotel'] == 'City Hotel') & (df['is_canceled'] == '0')]
```

    month column is not in order, if we visualize it we will not going to get good insights.<br>

```python
import sort_dataframeby_monthorweek as sd
def sort_month(df, column):
    return sd.Sort_Dataframeby_Month(df, column)
```

```python
hotel_prices = sort_month(hotel, 'month')
hotel_prices
```

   - `August` has the highest average price for both types of hotels, with Resort Hotels reaching `181.21` and City Hotels `118.41`. This is likely due to increased demand during the vacation season.

2. **Price Differences Between Hotel Types**:
   - `Resort Hotels` generally have a higher price than `City Hotels` in most months, especially from `June` to `September`. This could indicate that Resort Hotels experience higher demand or cater to more premium offerings during these months.
   - In the off-season months, such as `January`, `February`, and `November`, the price difference between Resort and City Hotels is much smaller, suggesting a more balanced demand during these months.

   - `January` and `November` have the lowest prices for both Resort and City Hotels, which aligns with typical low-season trends when tourism demand is generally lower.
   - `Resort Hotel` prices are consistently lower than City Hotels in January and November, indicating that resorts may have more flexibility in lowering prices to attract guests during off-peak times.
